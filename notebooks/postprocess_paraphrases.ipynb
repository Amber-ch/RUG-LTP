```python
training = pd.read_csv("training_paraphrased.csv")
test = pd.read_csv("test_paraphrased.csv")
validation = pd.read_csv("validation_paraphrased.csv")
```

```python
test
```

```python
def split_df(df):
  p_0 = df[["paraphrase_0", "Argument ID", "Conclusion", "Stance"]]
  p_0["Paraphrase Argument ID"] = p_0["Argument ID"] + "_0"
  p_0.rename(columns={"paraphrase_0":"Premise"}, inplace=True)

  p_1 = df[["paraphrase_1", "Argument ID", "Conclusion", "Stance"]]
  p_1["Paraphrase Argument ID"] = p_1["Argument ID"] + "_1"
  p_1.rename(columns={"paraphrase_1":"Premise"}, inplace=True)

  p_2 = df[["paraphrase_2", "Argument ID", "Conclusion", "Stance"]]
  p_2["Paraphrase Argument ID"] = p_2["Argument ID"] + "_2"
  p_2.rename(columns={"paraphrase_2":"Premise"}, inplace=True)

  p_3 = df[["paraphrase_3", "Argument ID", "Conclusion", "Stance"]]
  p_3["Paraphrase Argument ID"] = p_3["Argument ID"] + "_3"
  p_3.rename(columns={"paraphrase_3":"Premise"}, inplace=True)

  p_4 = df[["paraphrase_4", "Argument ID", "Conclusion", "Stance"]]
  p_4["Paraphrase Argument ID"] = p_4["Argument ID"] + "_4"
  p_4.rename(columns={"paraphrase_4":"Premise"}, inplace=True)

  return p_0, p_1, p_2, p_3, p_4
    
```

```python
test0, test1, test2, test3, test4 = split_df(test)
train0, train1, train2, train3, train4 = split_df(training)
val0, val1, val2, val3, val4 = split_df(validation)
```

```python
test_paraphrases = pd.concat([test0, test1, test2, test3, test4])
training_paraphrases = pd.concat([train0, train1, train2, train3, train4])
validation_paraphrases = pd.concat([val0, val1, val2, val3, val4])
```
